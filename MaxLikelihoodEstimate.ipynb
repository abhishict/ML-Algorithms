```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import linalg as LA
import numpy.random as nrm
import math
```

Write a program that finds the maximum-likelihood estimates (MLE) for the parameters of the following distributions:
(a) Binomial,
(b) Poisson,
(c) Exponential,
(d) Gaussian,
(e) Laplacian.

```python
def exponential(x,n,bins):
    theta = np.mean(x)
    print("theta :",theta)
    y = nrm.exponential(theta,n)
    fig, (ax1,ax2) = plt.subplots(1,2)
    ax1.hist(x,bins,ec='k',fc='k')
    ax2.hist(y,bins,ec='k',fc='k')
    ax1.set_title("Ground Truth")
    ax2.set_title("Estimated")    
    fig.suptitle("N = %d"%n)
```

```python
n1 = 1000
n2 = 10000
bins = 30
b = 5
x1 = np.random.exponential(b,n1)
x2 = np.random.exponential(b,n2)
print("\t \t MLE - EXPONENTIAL")
exponential(x1,n1,bins)
exponential(x2,n2,bins)
```

```python
def poisson(x,n,bins):
    a = np.mean(x)
    print("lam :",a)
    y = nrm.poisson(a,n)
    fig, (ax1,ax2) = plt.subplots(1,2)
    ax1.hist(x,bins,ec='k',fc='k')
    ax2.hist(y,bins,ec='k',fc='k')
    ax1.set_title("Ground Truth")
    ax2.set_title("Estimated")    
    fig.suptitle("N = %d"%n)
```

```python
n1= 1000
n2 = 10000
bins = 30
b = 6
x1 = np.random.poisson(b,n1)
x2 = np.random.poisson(b,n2)
print("\t \t MLE - POISSON")
poisson(x1,n1,bins)
poisson(x2,n2,bins)
```

```python
def gaussian(x,n,bins):
    b = np.std(x)
    a = np.mean(x)
    print("mu:",a)
    print("sigma:",b)
    y = nrm.normal(a,b,n)
    fig, (ax1,ax2) = plt.subplots(1,2)
    ax1.hist(x,bins,ec='k',fc='k')
    ax2.hist(y,bins,ec='k',fc='k')
    ax1.set_title("Ground Truth")
    ax2.set_title("Estimated")    
    fig.suptitle("N = %d"%n)   
```

```python
n1= 1000
n2 = 10000
bins = 30
mu = 5
sig = 2
x1 = np.random.normal(mu,sig,n1)
x2 = np.random.normal(mu,sig,n2)
print("\t \t MLE - GAUSSIAN")
gaussian(x1,n1,bins)
gaussian(x2,n2,bins)
```

```python
def binomial(x,n,m,bins):
    p= np.mean(x)/m
    print("n :", m)
    print("p ;",p)
    y = nrm.binomial(m,p,n)
    fig, (ax1,ax2) = plt.subplots(1,2)
    ax1.hist(x,bins,ec='k',fc='k')
    ax2.hist(y,bins,ec='k',fc='k')
    ax1.set_title("Ground Truth")
    ax2.set_title("Estimated")    
    fig.suptitle("N = %d"%n)
```

```python
n1= 1000
n2 = 10000
bins = 50
m = 60
v = 0.5
x1 = np.random.binomial(m,v,n1)
x2 = np.random.binomial(m,v,n2)
print("\t \t MLE - BINOMIAL")
binomial(x1,n1,m,bins)
binomial(x2,n2,m,bins)
```

```python
def laplacian(x,n,bins):
    a = np.mean(x)
    b = np.median(x)
    u = abs(np.array(x)-a)
    u = sum(u)/n
    c = math.sqrt(2)*u
    print("loc :",b)
    print("scale:",c)
    y = nrm.laplace(b,c,n)
    fig, (ax1,ax2) = plt.subplots(1,2)
    ax1.hist(x,bins,ec='k',fc='k')
    ax2.hist(y,bins,ec='k',fc='k')
    ax1.set_title("Ground Truth")
    ax2.set_title("Estimated")    
    fig.suptitle("N = %d"%n)   
```

```python
n1= 1000
n2 = 10000
bins = 50
m = 0
v = 4
x1 = np.random.laplace(m,v,n1)
x2 = np.random.laplace(m,v,n2)
print("\t \t MLE - LAPLACIAN")
laplacian(x1,n1,bins)
laplacian(x2,n2,bins)
```
